```python
from zipfile import ZipFile
from urllib.request import urlopen   
import pandas as pd
import os

URLs = ['https://s3.amazonaws.com/tripdata/201801-citibike-tripdata.csv.zip',\
       'https://s3.amazonaws.com/tripdata/201802-citibike-tripdata.csv.zip',\
       'https://s3.amazonaws.com/tripdata/201803-citibike-tripdata.csv.zip',\
       'https://s3.amazonaws.com/tripdata/201804-citibike-tripdata.csv.zip',\
       'https://s3.amazonaws.com/tripdata/201804-citibike-tripdata.csv.zip',\
       'https://s3.amazonaws.com/tripdata/201806-citibike-tripdata.csv.zip',\
       'https://s3.amazonaws.com/tripdata/201807-citibike-tripdata.csv.zip',\
       'https://s3.amazonaws.com/tripdata/201808-citibike-tripdata.csv.zip',\
       'https://s3.amazonaws.com/tripdata/201809-citibike-tripdata.csv.zip',\
       'https://s3.amazonaws.com/tripdata/2018010-citibike-tripdata.csv.zip',\
       'https://s3.amazonaws.com/tripdata/2018011-citibike-tripdata.csv.zip',\
       'https://s3.amazonaws.com/tripdata/2018012-citibike-tripdata.csv.zip']
       
# open and save the zip file onto computer
dfs=[]
for URL in URLs:
# open and save the zip file onto computer
    url = urlopen(URL)
    output = open('zipFile.zip', 'wb')    # note the flag:  "wb"        
    output.write(url.read())
    output.close()

# read the zip file as a pandas dataframe
    df = pd.read_csv('zipFile.zip')      
    dfs.append(df)
# if keeping on disk the zip file is not wanted, then:
#os.remove(201307-citibike-tripdata.zip)   # remove the copy of the zipfile on disk
```
